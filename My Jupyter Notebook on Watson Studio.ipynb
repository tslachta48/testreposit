### The following code illustrates a Pythagorean Theorem equation - $a⌃{2}$ + $b⌃{2}$ = $c⌃{2}$

```python
import math
def pythag(a, b):
    sum_of_squares = a * a + b * b
    c = math.sqrt(sum_of_squares)
    
    return c

solution = pythag(3, 4)

print(solution)
```
